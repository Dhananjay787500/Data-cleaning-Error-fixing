### Problem Restatement

Given a linked list, repeatedly take:

first node

last node

Add their values and track the maximum pair sum.

### Optimal Approach (O(1) extra space)

Exactly what you described:

Use fast & slow pointers to find the middle

Reverse second half of the list

Traverse both halves and compute max pair sum

```python
# Solution
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def pairSum(head: ListNode) -> int:
    # Step 1: Find the middle
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Step 2: Reverse the second half
    prev = None
    while slow:
        nxt = slow.next
        slow.next = prev
        prev = slow
        slow = nxt

    # Step 3: Compute max twin sum
    max_sum = 0
    left, right = head, prev
    while right:
        max_sum = max(max_sum, left.val + right.val)
        left = left.next
        right = right.next
    return max_sum
```
